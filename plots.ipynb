```python
import matplotlib.pyplot as plt
import numpy as np
```

1.
Basic Line Plot
o
Plot a simple line graph of ùë¶=ùë•2 for ùë• values ranging from -10 to 10.
o
Label the x-axis as "X Values" and the y-axis as "Squared Values."
o
Add a title to the graph.

```python
x = np.arange(-10, 11, 1)
y = x ** 2
plt.plot(x, y, label='y = x¬≤')
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Line Plot of y = x¬≤')
plt.grid(True)
plt.legend()
plt.show()
```

2.
Bar Plot
o
Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi', 'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
o
Add labels to the x-axis and y-axis and give the chart a title.

3.
Scatter Plot
o
Generate two lists of 20 random integers each (for x and y coordinates).
o
Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
import random
x = random.sample(range(1, 100), 20)
y = random.sample(range(1, 100), 20)
sizes = [random.randint(50, 300) for _ in range(20)]
plt.scatter(x, y, c='blue', s=sizes, alpha=0.6, edgecolors='w', label='Random Points')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot with Random Values')
plt.legend()
plt.show()
```

4.
Histogram
o
Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.
o
Customize the number of bins and add a grid to the plot.

```python
data = np.random.randn(100)
plt.hist(data, bins=10, color='purple', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Normal Data')
plt.grid(True)
plt.show()
```

5.
Pie Chart
o
Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
o
Display the percentage values on the chart and use a color palette for differentiation.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['gold', 'lightblue', 'lightgreen', 'pink', 'orange']
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Percentage Distribution of Categories')
plt.show()
```

6.

o
Create a box plot to visualize the distribution of the marks scored by students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
